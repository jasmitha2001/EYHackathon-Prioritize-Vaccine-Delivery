## Reading csv file

```python
covid=pd.read_csv("state_wise.csv") #state wise covid dataset
```

```python
covid
```

```python
covid.columns #Columns in covid dataset
```

```python
covid = covid[['State', 'Confirmed', 'Recovered', 'Deaths', 'Active']] ## retriving required attributes
```

```python
covid
```

```python
top5_Active=covid.sort_values(by="Active",ascending=False) # Soring based on active cases in decending order
```

## Plotting bargraph

### Top 5 states based on no of active cases

```python
sns.set(rc={'figure.figsize':(14,8)})
sns.barplot(x="State",y="Active",data=top5_Active[0:5],hue="State") 
plt.show()
```

```python
sns.set(rc={'figure.figsize':(14,8)})
sns.barplot(x="State",y="Deaths",data=top5_Deaths[0:5],hue="State")
plt.show()
```

```python
population=population.rename(columns={"Aadhaar assigned as of 2019": "POPULATION"}) ## Renaming column names
```

```python
population
```

```python
i = covid[covid.State  == 'Total'].index ## Selecting records based on condtion
covid=covid.drop(i) ## Dropping selected records from table
```

```python
covid
```

```python
covid19_data = pd.merge(covid, population, on='State') ## Merging two data sets using states
```

```python
covid19_data
```

```python
top1=covid19_data.sort_values(["Confirmed","POPULATION"], ascending = (False, False))
top1=top1[0:1]
top1
```

```python
confirmed=top1['Confirmed'].sum()
```

```python
population=top1['POPULATION'].sum()
```

```python
a=[confirmed,population]
fig = plt.figure(figsize =(15, 7)) 
plt.rcParams['font.size'] = 15
plt.pie(a,labels=['Confirmed','Population'],colors=['blue','red'],autopct='%1.1f%%')
plt.axis("equal")
plt.title("Population density")
```

```python
patients_details = pd.read_csv("patients_data.csv") ## Retriving covid patients dataset
```

```python
patients_details.columns
```

```python
patients_details = patients_details[['Age Bracket','Gender', 'Detected City',
       'Detected District', 'Detected State', 'Backup Notes']] ## selecting required columns
```

```python
patients_details=patients_details.rename(columns={"Detected District": "District"}) # Renaming column names
patients_details=patients_details.rename(columns={"Detected State": "State"})
```

```python
patients_details
```

```python
patients_details=patients_details.dropna() 
```

```python
patients_details=pd.merge(covid19_data, patients_details, on='State') ##Merging two data sets using states
patients_details
```

```python
males=(patients_details['Gender']=='M').sum() # finding number of males under covid 19 
females=(patients_details['Gender']=='F').sum() # finding number of males under covid 19
males
```

## Pie chart

## Pie chart representing number of males and females

```python
a=[males,females]
fig = plt.figure(figsize =(10, 7)) 
plt.rcParams['font.size'] = 20
plt.pie(a,labels=['Males','Females'],colors=['blue','red'],autopct='%1.1f%%')
plt.axis("equal")
plt.title("Gender Proportion")
```

```python
data_types_dict = {'Age Bracket': int} 
patients_details = patients_details.astype(data_types_dict)
```

```python
age1=((patients_details['Age Bracket']>=1) & (patients_details['Age Bracket']<15)).sum()
age2=((patients_details['Age Bracket']>=15) & (patients_details['Age Bracket']<30)).sum()
age3=((patients_details['Age Bracket']>=30) & (patients_details['Age Bracket']<50)).sum()
age4=((patients_details['Age Bracket']>=50) & (patients_details['Age Bracket']<75)).sum()
age5=((patients_details['Age Bracket']>=75)).sum()
```

```python
ag=[age1,age2,age3,age4,age5]
```

```python
fig = plt.figure(figsize =(10, 7)) 
plt.rcParams['font.size'] = 17
plt.pie(ag,labels=['0-15','15-30','30-50','50-75','>75'],colors=['blue','red','green','yellow','pink'],autopct='%1.1f%%')
plt.axis("equal")
plt.title("Age Proportion")
```
